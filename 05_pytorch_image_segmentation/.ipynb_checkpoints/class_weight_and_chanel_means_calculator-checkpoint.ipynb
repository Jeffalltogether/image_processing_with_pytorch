```python
# define some transformations on numpy arrays we will use for image augmentations
def Random_Sized_Crop(img, mask, width=224, height=224):
    import random

    assert img.shape[0] == mask.shape[0]
    assert img.shape[1] == mask.shape[1]

    if img.shape[0] < height:
        img = imresize(img, (height,img.shape[1],img.shape[2]), 'bilinear')
        mask = imresize(mask, (height,img.shape[1]), 'nearest')        
        
    if img.shape[1] < width:
        img = imresize(img, (img.shape[0],width,img.shape[2]), 'bilinear')
        mask = imresize(mask, (img.shape[0],width), 'nearest')   
        
    x = random.randint(0, img.shape[1] - width)
    y = random.randint(0, img.shape[0] - height)
    img = img[y:y+height, x:x+width]
    mask = mask[y:y+height, x:x+width]

    return img, mask

def Random_Horizantal_Flip(img, mask):
    import numpy as np

    r_num = np.round(np.random.rand(1))
    if r_num == 0.0:
        img = np.fliplr(img).copy()
        mask = np.fliplr(mask).copy()

    return img, mask
```

```python
class COCO_segmentation_helper(object):
    '''
    Notes
    '''
    def __init__(self, dataDir, dataType, catIds = ['zebra'], RandomSizedCrop = 224):
        '''
        Notes
        '''

        # define COCO directory, data type, and categories
        self.dataDir = dataDir
        self.dataType = dataType
        self.catIds = catIds
        self.ignore_label = len(catIds) + 1

        # Generate
        # make dictionary from classes desired by user
        self.label_dict = dict()
        for i,cat in enumerate(catIds):
            self.label_dict.update({cat:i+1})
        print self.label_dict

        # initialize COCO api for testing instance annotations
        annFile = '{}/annotations/instances_{}.json'.format(self.dataDir, self.dataType)
        self.coco = COCO(annFile)

    def __call__(self, im_list, im_width = 224, im_height = 224, batch_size = 1): #, im_Label):
        '''
        Notes
        '''
        # get the desired category IDs
        catIds = self.coco.getCatIds(catNms=self.catIds)
        
        # check that im_Number is a list
        if type(im_list) not in [list, 'numpy.ndarray']:
            "im_Number must be list or numpy.ndarray format"
            return
        
        # create empty tensors to be filled with color images
        batch_images = np.zeros([batch_size, 3, im_width, im_height])
        batch_masks = np.zeros([batch_size, im_width, im_height])

        for i,im_Number in enumerate(im_list):
            # get image number
            im_Number = int(im_Number)

            # read the original image
            img = self.coco.loadImgs(im_Number)[0]
            PIL_image = Image.open(os.path.join(self.dataDir, 'images', self.dataType, img['file_name']))
            
            w, h = PIL_image.size
            data_arr = np.array(PIL_image)
            
            # generate the GT mask image
            mask = np.zeros((h, w))

            for cat,label in self.label_dict.items():
                catIds = self.coco.getCatIds(catNms=cat);
                annIds = self.coco.getAnnIds(imgIds=im_Number, catIds=catIds, iscrowd=False)
                anns = self.coco.loadAnns(annIds)
                for ann in anns:
                    gt = self.coco.annToMask(ann)
                    mask[gt == 1] = label

            # augment image
            im_data, mask = Random_Sized_Crop(data_arr, mask, im_width, im_height)
            im_data, mask = Random_Horizantal_Flip(im_data, mask)

            im_data = np.swapaxes(im_data,0,2)
            batch_images[i,:,:,:] = im_data
            batch_masks[i,:,:] = mask

        return {'image': batch_images, 'label': batch_masks}

    def get_classifier_data(self):
        '''
        After initialization, the function get_classifier_data is called to generate the dataframe of images that
        contain the classes desired by user
        '''
        # get list of images that contain the desired categories
        data = set()
        for cat in self.catIds:
            classId = self.coco.getCatIds(catNms=cat)
            imageIds = self.coco.getImgIds(catIds=classId)
            data.update(imageIds)

        # convert to dataframe and add labels
        data = np.array(list(data), dtype=str)

        # shuffle before returning
        np.random.shuffle(data)
        self.shuffled_data = data

        return self.shuffled_data
```

```python
# Define the classes to include in the classifier
classes = ['cake', 'zebra']
num_classes = len(classes) + 2 # n + 2 to include background class and ignore class

# generate training and test data from the COCO dataset
training_read_data = COCO_segmentation_helper(dataDir, 'train2017', catIds = classes) # define the parameters for read image function
testing_read_data = COCO_segmentation_helper(dataDir, 'val2017', catIds = classes)

training_data = training_read_data.get_classifier_data()  #[:3000] # get first 200 images
testing_data = testing_read_data.get_classifier_data()[:15] # get first 15 images

print training_data[:15]
print testing_data[:15]
```

```python
# find B&W images and Images where mask dims != iamge dims, then remove from training and validation images
def find_bad_images(image_list, dataDir, dataType, classes, coco_reader):
    bad_images = []
    for im_Number in image_list:
        # get image number
        im_Number = int(im_Number)

        # read the original image
        img = coco_reader.loadImgs(im_Number)[0]
        PIL_image = Image.open(os.path.join(dataDir, 'images', dataType, img['file_name']))
        data_arr = np.array(PIL_image)
        
        # read instance annotation
        mask = []
        for cat in classes:
            catIds = coco_train.getCatIds(catNms=cat);
            annIds = coco_train.getAnnIds(imgIds=[im_Number], catIds=catIds, iscrowd=False)
            anns = coco_train.loadAnns(annIds)
            for ann in anns:
                mask = coco_train.annToMask(ann)

        # some images have no postivie targets
        if isinstance(mask, list):
            print 'bad mask {}'.format(im_Number)
            bad_images.append(str(im_Number))
        
        # check that the image and mask have the same dimensions
        elif (mask.shape[:2] != data_arr.shape[:2]):
            print 'mask shape != image shape {}'.format(im_Number)            
            bad_images.append(str(im_Number))

        # some B&W images have onle 1 channel            
        elif len(data_arr.shape) != 3:
            print 'black and white image {}'.format(im_Number)
            bad_images.append(str(im_Number))
            
        # some B&W images have 3 channels, but they are all equivalent
        elif np.array_equal(data_arr[:,:,0],data_arr[:,:,1]):
            print 'black and white image {}'.format(im_Number)
            bad_images.append(str(im_Number))

        else:
            continue
        
    return bad_images

training_bad_images = find_bad_images(training_data, dataDir, 'train2017', classes, coco_train)
training_data = [im for im in training_data if im not in training_bad_images]

testing_bad_images = find_bad_images(testing_data, dataDir, 'val2017', classes, coco_test)
testing_data = [im for im in testing_data if im not in testing_bad_images]
```

## add identification of channel means 

```python
all_classes = classes+['background']
class_representation = {c:0 for c in all_classes}
channel_means = {'channel_1':[0.0,0.0], 'channel_2':[0.0,0.0], 'channel_3':[0.0,0.0]}
num_epochs = 1
batch_size = 112

for n,epoch in enumerate(range(num_epochs)):  # again, normally you would NOT do 300 epochs, it is toy data
    # notify user of epoch
    print 'Starting Epoch # %i' %(n+1)

    batch_channel_means = []
    #iterate though data
    for b,batch_num in enumerate(range(len(training_data)/batch_size)):
        # get lower and upper bound domain of batch
        batch_lower, batch_upper = batch_num*batch_size, (batch_num*batch_size) + batch_size
        
        # get image and label data for the batch
        im_Number = training_data[batch_lower:batch_upper]
        im_Number = list(im_Number)
        data_dict = training_read_data(im_Number, batch_size = batch_size)

        in_img = data_dict['image']
        in_label = data_dict['label']

        # sum label pixels to obtain class representations
        class_representation['background'] += in_label.shape[0]*in_label.shape[1]*in_label.shape[2]
        for key,value in training_read_data.label_dict.items():
            num_pixels = len(in_label[in_label==value])
            class_representation[key] += num_pixels
            class_representation['background'] -= num_pixels
        
        # average channel values
        b = np.mean(in_img, axis=len(in_img.shape)-1)
        c = np.mean(b, axis=len(b.shape)-1)
        d = np.mean(c, axis=len(c.shape)-2)
        batch_channel_means.append(d)
        
    print class_representation
    for key,value in class_representation.items():
        if value != 0.0:
            print '\nThe class weight for {} is {} \n'.format(key, float(class_representation['background']) / value)
    print 'The order of the weight vector is: {}\n'.format(training_read_data.label_dict)
    
channel_1, channel_2, channel_3 = [], [], []
for i,j,k in batch_channel_means:
    channel_1.append(i)
    channel_2.append(j)
    channel_3.append(k)

channel_means['channel_1'] = [np.mean(channel_1), np.std(channel_1)]
channel_means['channel_2'] = [np.mean(channel_2), np.std(channel_2)]
channel_means['channel_3'] = [np.mean(channel_3), np.std(channel_3)]
```
